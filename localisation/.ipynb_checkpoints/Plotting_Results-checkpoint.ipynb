```python
def parse_dbow2_results(loc, db_paths, query_paths):

    rs = {}
    with open(loc) as f:
        line = ""
        while not line.startswith('Searching'):
            line = next(f)
        while True:
            qn = int(line.split(' ')[3][:-1])
            rfq = []
            for _ in range(100):
                rfq.append(int(next(f).split(' ')[1][:-1]))
            rs[qn] = rfq
            line = next(f)
            if line.startswith('n images::'):
                break
            assert line.startswith('Searching')
    
    results = {}
    for qn, rns in rs.items():
        query_path = query_paths[qn]
        result_paths = [db_paths[rn] for rn in rns]
        results[query_path] = result_paths
        
    return results

def process_log_file(loc):
    records = []
    with open(loc) as f:
        for line in f:
             records.append(json.loads(line))

    val_res = []
    val_records = []
    losses = []
    val_losses = []
    for record in records:
        losses.append(record['loss'])
        if 'dubrovnik_results' in record:
            ps = list(sorted([(int(k), v) for k, v in record['dubrovnik_results'].items()]))
            ps = [[-1, ps[0][1]], [0, ps[0][1]]] + ps
            ps = np.array(ps)
            val_res.append(ps)
            val_records.append(record)
        if 'val_loss' in record:
            val_losses.append([record['epoch_n'], record['val_loss']])
    val_losses = np.array(val_losses)

    plt.plot(*val_losses.T)
    minimiser = np.argmin(val_losses[:, 1])
    plt.axvline(minimiser, linestyle='--')
    print(minimiser)
    return val_res[minimiser]
```

```python
selected_results['MobileNet FineTuned 128D Mean Pooling'] = process_log_file('/home/nm583/sfm_data/log.ndjson')
```

```python
baseline_mobilenet = {}
with open('/home/nm583/sfm_data/raw_mobilenet_eval') as f:
    for line in f:
        line = line.strip()
        if line.startswith('JSON'):
            d = json.loads(line[len('JSON'):])
            baseline_mobilenet[d[0]] = d[1]
            
for k in baseline_mobilenet:
    v = list(sorted([(int(a), b) for a, b in baseline_mobilenet[k].items()]))
    v = [[-1, v[0][1]], [0, v[0][1]]] + v
    baseline_mobilenet[k] = np.array(v)
    
selected_results['MobileNet 1024D Mean Pooling'] = baseline_mobilenet['mean']
selected_results['MobileNet 1024D Max Pooling'] = baseline_mobilenet['max']
```
